```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.info()
```

```python
# find missing values
titanic.isnull().sum()
```

```python
# Plots
titanic.survived.value_counts().plot(kind='bar') 
```

```python
sns.countplot(x="survived",hue='sex', data=titanic)
```

```python
sns.scatterplot(x='age', y='fare',hue="pclass",data=titanic)
```

```python
plt.scatter(x='age', y='fare',data=titanic)
```

```python
sns.barplot(x='sex', y='fare',hue="pclass",data=titanic)
```

```python
sns.pairplot(titanic)
```

```python
titanic.hist(bins=10)
```

```python
sns.boxplot(data=titanic, x='sex',y='fare',hue='class')
```

```python
sns.boxenplot(data=titanic, x='sex',y='fare',hue='class')
plt.title('titanic ka data')
plt.xlabel('x_label_sex')
plt.ylabel('y_label_fare')
```

```python
sns.stripplot(data=titanic, x='sex',y='fare',hue='class')
plt.title('titanic ka data')
plt.xlabel('x_label_sex')
plt.ylabel('y_label_fare')
```

```python
corr=titanic.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
```

```python
sns.violinplot(data=titanic, x='sex',y='fare',hue='class')
plt.title('titanic ka data')
plt.xlabel('x_label_sex')
plt.ylabel('y_label_fare')
```
