# Import Libraries
## seaborn, matplotlib, pandas,numpy,scipy


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.lineplot(data=iris, x ="sepal_length", y ="sepal_width")
plt.show()
```

```python
# Title
sns.lineplot(data=iris, x ="sepal_length", y ="sepal_width")
plt.title("iris data")
plt.show()
```

```python
# Adding Limit
sns.lineplot(data=iris, x ="sepal_length", y ="sepal_width")
plt.title("iris data")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

```python
# Set style darkgrid, whitegrid, dark, white, ticks
sns.lineplot(data=iris, x ="sepal_length", y ="sepal_width")
plt.title("iris data")
sns.set_style(style=None, rc = None)
plt.show()
```

```python
# Size of figure
plt.figure(figsize=(10,8))
sns.lineplot(data=iris, x ="sepal_length", y ="sepal_width")
plt.title("iris data")
plt.show()
```

```python
# Bar plot
sns.barplot(data=iris, x ="species", y ="sepal_width")
plt.title("iris data")
plt.show()
```

```python
sns.barplot(data=iris, x ="species", y ="petal_length")
plt.title("iris data")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.barplot(data=titanic, x ="sex", y ="alone", hue = "who")
plt.title("Titanic Data")
plt.show()
```

```python
sns.barplot(data=titanic, order=["female","male"], x ="sex", y ="alone", hue = "who", color = "red")
plt.title("Titanic Data")
plt.show()
```

```python
sns.barplot(data=titanic, order=["female","male"], x ="sex", y ="alone", hue = "who", color = "red", ci= None)
plt.title("Titanic Data")
plt.show()
```

```python
sns.barplot(data=titanic, order=["female","male"], x ="sex", y ="alone", 
            hue = "who", color = "red", ci= None, palette="pastel")
plt.title("Titanic Data")
plt.show()
```

```python
sns.barplot(data=titanic,  x ="sex", y ="fare", 
            hue = "who", estimator="mean")
plt.title("Titanic Data")
plt.show()
```

```python
# Data Visual 2nd Part
sns.barplot(data=titanic,  x ="fare", y ="class", 
            hue = "sex", color= "red", estimator="mean")
plt.title("Titanic Data")
plt.show()
```

```python
sns.barplot(data=titanic,  x ="class", y ="fare", 
            linewidth=2.5, facecolor=(1,1,1,0),
            errcolor=".2", edgecolor=".2")
plt.title("Titanic Data")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=titanic,  x ="class", y ="fare")
plt.title("Titanic Data")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=titanic,  x ="who", y ="fare")
plt.title("Titanic Data")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y =titanic["age"])
plt.title("Titanic Data")
```

```python
sns.boxplot(data=titanic,  x ="survived", y ="age")
plt.title("Titanic Data")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=titanic,  x ="survived", y ="age", showmeans=True)
plt.title("Titanic Data")
plt.show()
```
